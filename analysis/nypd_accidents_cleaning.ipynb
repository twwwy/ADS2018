```python
import numpy as np
import urllib.request
import zipfile
import pandas as pd
```

```python
urllib.request.urlretrieve("https://data.cityofnewyork.us/api/views/h9gi-nx95/rows.csv?accessType=DOWNLOAD","accidents.csv")
```

```python
accidents = pd.read_csv("accidents.csv")
```

```python
accidents.head()
```

```python
accidents['DATE'] = pd.to_datetime(accidents['DATE'])
```

```python
accidents.head()
```

```python
accidents = accidents[(accidents['DATE']>="2016-07-01") & (accidents['DATE']<"2018-09-01")]
```

```python
accidents.sort_values(by='DATE',ascending=False)
```

```python
accidents.columns
```

```python
accidents.columns = map(str.lower, accidents.columns)
accidents.columns = accidents.columns.str.replace(' ', '-')
```

```python
accidents.columns
```

```python
accidents.dropna(subset=['location'],inplace=True)
```

```python
accidents.head()
```

```python
accidents.to_csv("accidents_cleaned.csv")
```
