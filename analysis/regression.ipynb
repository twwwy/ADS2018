```python
import pandas as pd
import statsmodels.formula.api as smf
```

```python
df = pd.read_csv('accidents_cleaned.csv')
```

```python
df.head()
```

```python
injured = smf.ols(formula='number_of_persons_injured ~ distraction + driver_behavior + \
                  environment + vehicle_issue + fault_by_others + \
                  influence + driver_health + month + dayofweek + hour + boro_name', data=df).fit()
injured.summary()
```

```python
killed = smf.ols(formula='number_of_persons_killed ~ distraction + driver_behavior + \
                  environment + vehicle_issue + fault_by_others + \
                  influence + driver_health + month + dayofweek + hour + boro_name', data=df).fit()
killed.summary()
```
