```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sma
plt.style.use('seaborn')
%matplotlib inline
```

```python
df = pd.read_csv('data/cleaned.csv')
```

# OLS

```python
killed_mod = smf.ols(formula='number_of_persons_killed ~ C(contributing_factor_vehicle_1)', data=df).fit()
```

```python
killed_mod.summary()
```

# Time Serires Analysis

```python
df.head()
```

```python
df['datetime'] = pd.to_datetime(df.date.astype(str) + ' '  + df.time.astype(str))
```

```python
df.plot('datetime', 'number_of_persons_injured', kind='line')
```

```python
date_gb = df.groupby(df.date)['time'].count()
```

```python
date_gb.columns = ['accident_counts']
```

```python
date_gb.plot()
```

```python
#remove missing number dates
date_gb = date_gb.iloc[5:]
```

```python
date_gb.plot()
ax = date_gb.plot()
ax.set_xticklabels(df.date, rotation=45)
```

```python
# Decompose into cyclical and trend components.

cycle, trend = sma.tsa.filters.hpfilter(date_gb, lamb = 6.25)
```

```python
ax = cycle.plot()
ax.set_xticklabels(df.index, rotation=45)
```

```python
ax = trend.plot()
ax.set_xticklabels(df.index, rotation=45)
```
